#Features categóricas

##4.Tipo

```python
#@title 4.1 Tipo Gráfico Pizza

dados_tipo = df['tipo'].value_counts()

# Ordenar os valores e rótulos
dados_tipo_sorted = dados_tipo.sort_values(ascending=False)
labels = dados_tipo_sorted.index
sizes = dados_tipo_sorted.values

# Definir cores para as fatias do gráfico de pizza
colors = ['yellow', 'orange', 'green', 'red', 'purple', 'brown', 'blue']

# Reorganizar os valores e rótulos
new_order = [0, 4, 2, 3, 1, 5]  # Nova ordem
labels = labels[new_order]
sizes = sizes[new_order]
colors = [colors[i] for i in new_order]

# Criar o gráfico de pizza
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição dos Tipos de Veículos')
plt.axis('equal')  # Para garantir que o gráfico de pizza seja desenhado como um círculo

# Salvar o gráfico em formato PNG
#plt.savefig('meu_grafico_tipo.png', dpi=300)

# Exibir o gráfico de pizza
plt.show()
```

```python
#@title 4.2 Gráfico de Contagem
dados_tipo = df['tipo'].value_counts()

# Definir rótulos e contagens para o gráfico de contagem
labels = dados_tipo.index
counts = dados_tipo.values

# Definir posições das barras
x_pos = range(len(labels))

# Criar o gráfico de contagem
plt.figure(figsize=(8, 6))
plt.bar(x_pos, counts, align='center', alpha=0.8)
plt.xticks(x_pos, labels)
plt.xlabel('Tipo')
plt.ylabel('Contagem')
plt.title('Contagem dos Tipos de Veículos')
plt.xticks(rotation=45)

# Adicionar o número exato de tipos contados ao lado das barras
for i, count in enumerate(counts):
    plt.text(x_pos[i], count + 1, str(count), ha='center', va='bottom')

# Salvar o gráfico em formato PNG
#plt.savefig('meu_grafico_tipo.png', dpi=300)

# Exibir o gráfico de contagem
plt.show()
```

##5.Marca

```python
#@title 5.1 Marca Porcentagem da Participação
dados_tipo = df['marca'].value_counts()

# Calcular as porcentagens de participação
total = dados_tipo.sum()
porcentagens = (dados_tipo / total) * 100

# Definir rótulos e porcentagens para o gráfico de barras
labels = dados_tipo.index
percentages = porcentagens.values

# Definir posições das barras
y_pos = range(len(labels))

# Criar o gráfico de barras com eixos invertidos
plt.figure(figsize=(8, 6))
plt.barh(y_pos, percentages, align='center', alpha=0.8)
plt.yticks(y_pos, labels)  # Inverter os rótulos e o eixo y

plt.ylabel('Marca', fontsize=12)  # Definir o tamanho da fonte do rótulo do eixo y
plt.xlabel('Porcentagem de Participação', fontsize=12)  # Definir o tamanho da fonte do rótulo do eixo x
plt.title('Participação das Marcas de Veículos', fontsize=14)

# Adicionar as porcentagens exatas ao lado das barras com tamanhos de fonte menores
for i, percentage in enumerate(percentages):
    plt.text(percentage + 1, y_pos[i], f'{percentage:.1f}%', ha='left', va='center', fontsize=8)

# Salvar o gráfico em formato PNG
#plt.savefig('meu_grafico_marca.png', dpi=300)

# Exibir o gráfico de barras
plt.tight_layout()
plt.show()
```

```python
#@title 5.2 Contagem da Participação

dados_tipo = df['marca'].value_counts()

# Definir rótulos e contagens para o gráfico de contagem
labels = dados_tipo.index
counts = dados_tipo.values

# Definir posições das barras
y_pos = range(len(labels))

# Criar o gráfico de contagem com eixos invertidos
plt.figure(figsize=(8, 6))
plt.barh(y_pos, counts, align='center', alpha=0.8)
plt.yticks(y_pos, labels)  # Inverter os rótulos e o eixo y

plt.title('Contagem da Participação das Marcas')

# Adicionar o número exato de tipos contados ao lado das barras
for i, count in enumerate(counts):
    plt.text(count + 1, y_pos[i], str(count), ha='left', va='center', fontsize=8)

# Salvar o gráfico em formato PNG
#plt.savefig('meu_grafico_marca.png', bbox_inches='tight')

# Exibir o gráfico de contagem
plt.tight_layout()
plt.show()
```

##b.Pergunta
Qual o melhor estado para se comprar uma picape com transmissão automática e por quê?

```python

# Filtrar os carros com a palavra "AUTOMÁTICO" na coluna "versao"
df_automaticos = df[df['versao'].str.contains('AUTOMÁTICO', case=False, na=False)]

# Exibir o novo DataFrame
df_automaticos.head(10)
```

```python

# Filtrar os dados para incluir apenas as picapes
picapes_automaticas = df_automaticos[df_automaticos['tipo'] == 'Picape']

# Calcular a contagem de picapes automáticas por estado
contagem_por_estado = picapes_automaticas['estado_vendedor'].value_counts()

# Plotar o gráfico de barras
plt.figure(figsize=(6, 2))
contagem_por_estado.plot(kind='bar')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.title(' Picapes com Transmissão Automáticas por Estado')

# Salvar o gráfico em formato PNG
plt.savefig('meu_grafico_hipB.png', bbox_inches='tight')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python

# Agrupar os dados por estado e calcular a contagem de vendas por estado
contagem_vendas_por_estado = picapes_automaticas['estado_vendedor'].value_counts()

# Filtrar os estados com pelo menos 100 vendas
estados_com_100_vendas = contagem_vendas_por_estado[contagem_vendas_por_estado >= 100].index

# Calcular o preço médio por estado
preco_medio_por_estado = picapes_automaticas[picapes_automaticas['estado_vendedor'].isin(estados_com_100_vendas)].groupby('estado_vendedor')['preco'].mean()

preco_medio_por_estado = preco_medio_por_estado.round(2).astype(float)
print(preco_medio_por_estado )
```

```python

import numpy as np
import pickle  # salvar modelo
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Carregar os dados

df_xg = df # Passando o Data Frame

# Selecionar as colunas de interesse
columns = ['hodometro', 'ano_de_fabricacao', 'estado_vendedor', 'marca', 'tipo', 'cambio', 'anunciante', 'modelo',
           'versao', 'preco']
df_xg = df_xg[columns]

# Converter colunas para tipo categórico
categorical_columns = ['estado_vendedor', 'marca', 'tipo', 'cambio', 'anunciante', 'modelo', 'versao']
categorical_columns_indices = [df_xg.columns.get_loc(col) for col in categorical_columns]
df_xg.iloc[:, categorical_columns_indices] = df_xg.iloc[:, categorical_columns_indices].astype('category')

# Pré-processamento das colunas categóricas
encoder = OneHotEncoder(handle_unknown='ignore')
encoded_categorical = encoder.fit_transform(df_xg[categorical_columns]).toarray()
encoded_categorical_columns = encoder.get_feature_names_out(categorical_columns)

# Concatenar as colunas codificadas com as colunas numéricas
df_encoded = pd.concat([pd.DataFrame(encoded_categorical, columns=encoded_categorical_columns),
                        df_xg.drop(categorical_columns, axis=1)], axis=1)

# Pré-processamento das colunas numéricas
scaler = StandardScaler()
df_encoded['hodometro'] = scaler.fit_transform(df_encoded['hodometro'].values.reshape(-1, 1))

# Separar as features e o target
X = df_encoded.drop('preco', axis=1)
y = df_encoded['preco']

# Dividir o dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# Configurar o modelo XGBoost
model = xgb.XGBRegressor()

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer as previsões no conjunto de teste
y_pred = model.predict(X_test)

# Comparar os valores reais com os valores previstos
comparison = pd.DataFrame({'Real': y_test, 'Previsto': y_pred})
print(comparison)

# Calcular o erro médio quadrático (MSE)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Salvando o modelo
#with open('/content/previsao_carros_xgboost.pkl', 'wb') as file:
    #pickle.dump(model, file)

print("MSE:", mse)
print("RMSE:", rmse)
```
